# Questão 1 (vale até 2 pontos)
Determine a área entre as curvas $$y=xe^{-x^{2}}$$ $$y=x+1$$ $$x=2$$ e o eixo $y$, usando somente matplotlib, numpy, sympy. Trace o gráfico desta área preenchendo a área entre as curvas em uma cor diferente.



```python
# Importando as bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt
from sympy import *

# Definindo a variável simbólica x
x = symbols('x')

# Definindo as funções f(x) e g(x)
f = x*exp(-x**2)
g = x+1

integral_f = (integrate(f).subs(x, 2).evalf())-(integrate(f).subs(x, 0).evalf())
integral_g = (integrate(g).subs(x, 2).evalf())-(integrate(g).subs(x, 0).evalf())

print(f"Area entre as curvas:{integral_g-integral_f}")

# Converte as funções sympy para funções numpy
f = lambdify(x, f, "numpy")
g = lambdify(x, g, "numpy")

# Criando uma série de pontos x no intervalo de 0 a 3
x_vals = np.linspace(0, 3, 1000)

# Calculando os valores correspondentes de f(x) e g(x)
f_vals = f(x_vals)
g_vals = g(x_vals)

# Plotando a função f(x) e g(x) no mesmo gráfico e preenchendo a área entre as curvas
plt.figure(figsize=(8, 6))
plt.plot(x_vals, f_vals, label='$f(x) = xe^{-x^{2}}$')
plt.plot(x_vals, g_vals, label='$g(x) = -x+1$')
plt.fill_between(x_vals, f_vals, g_vals, where=(0<=x_vals) & (x_vals<=2), color='lightblue', alpha=0.5)
plt.axvline(0, color='r', linestyle='--')  # Linha pontilhada em x = 1
plt.axvline(2, color='r', linestyle='--')  # Linha pontilhada em x = 2
plt.title('Área entre as funções $f(x)$ e $g(x)$ entre $x=1$ e $x=2$')
plt.legend()
plt.grid()
plt.show()
```

# Questão 2 (vale até 2 pontos)
Resolva, algebricamente, explicitando em latex todos os passos necessários, as seguintes integrais:

##a) 
$$\int \cos(3t) \sin(8t) dt $$
##b)
$$\int 4x \cos(2-3x) dx$$



##a) 
Integracao por partes
$$u = \cos(3t), dv = \sin(8t), du = dt, v = \frac{-\cos(8t)}{8}$$

$$\int \cos(3t) \sin(8t) dt=\cos(3t) \frac{-\cos(8t)}{8}-\int (\frac{-\cos(8t)}{8})dt=\frac{-\cos(3t)\cos(8t)}{8}+\int \frac{\cos(8t)}{8}dt=\frac{-\cos(3t)\cos(8t)}{8}+ \frac{\sin(8t)}{64}+c$$
Resposta:
$$\int \cos(3t) \sin(8t) dt=\frac{-\cos(3t)\cos(8t)}{8}+ \frac{\sin(8t)}{64}+c$$

##b)
Integracao por partes
$$u = 4x, dv = \cos(2-3x), du = dx, v = \frac{-\sin(2-3x)}{3}$$

$$\int 4x \cos(2-3x) dx=4x \frac{-\sin(2-3x)}{3} - \int (\frac{-\sin(2-3x)}{3}) dx=\frac{-4x \sin(2-3x)}{3} + \int \frac{\sin(2-3x)}{3} dx=\frac{-4x \sin(2-3x)}{3} + \frac{\cos(2-3x)}{9}+c$$
Resposta:
$$\int 4x \cos(2-3x) dx=\frac{-4x \sin(2-3x)}{3} + \frac{\cos(2-3x)}{9}+c$$

# Questão 3 (vale até 4 pontos)
Calcule o valor das integrais definidas explicitadas a seguir. Trace o gráfico das funções que foram integradas e das funções encontradas após a integração destacando, com linhas verticais, os limites de integração. 

##a)
$$\int_0^2 x^2+1 \space dx$$

##b) 
$$\int_{-3}^1 6x^2-5x+2 \space dx$$



```python
# Importando as bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt
from sympy import *

# Definindo a variável simbólica x
x = symbols('x')

# Definindo as funções f(x) e g(x)
f = x**2+1
g = 6*x**2-5*x+2

integral_f = integrate(f)
integral_g = integrate(g)

print(f"valor de a: {(integral_f.subs(x, 2).evalf())-(integral_f.subs(x, 0).evalf())}")
print(f"valor de b: {(integral_g.subs(x, 1).evalf())-(integral_g.subs(x, -3).evalf())}")

# Converte as funções sympy para funções numpy
f = lambdify(x, f, "numpy")
g = lambdify(x, g, "numpy")
integral_f = lambdify(x, integral_f, "numpy")
integral_g = lambdify(x, integral_g, "numpy")

# Criando uma série de pontos x no intervalo de 0 a 3
x_vals = np.linspace(-4, 3, 1000)

# Calculando os valores correspondentes de f(x) e g(x)
f_vals = f(x_vals)
g_vals = g(x_vals)
integral_f_vals = integral_f(x_vals)
integral_g_vals = integral_g(x_vals)

# Plotando a função f(x) e g(x) no mesmo gráfico e preenchendo a área entre as curvas
plt.figure(figsize=(8, 6))
plt.plot(x_vals, f_vals, label='$x^2+1$')
plt.plot(x_vals, integral_f_vals, label='$\int_0^2 x^2+1dx$')
plt.plot(x_vals, g_vals, label='$6x^2-5x+2$')
plt.plot(x_vals, integral_g_vals, label='$\int_{-3}^1 6x^2-5x+2dx$')
plt.axvline(0, color='r', linestyle='--')  # Linha pontilhada em x = 0
plt.axvline(2, color='r', linestyle='--')  # Linha pontilhada em x = 2
plt.axvline(-3, color='b', linestyle='--')  # Linha pontilhada em x = 0
plt.axvline(1, color='b', linestyle='--')  # Linha pontilhada em x = 2
plt.legend()
plt.grid()
plt.show()
```
