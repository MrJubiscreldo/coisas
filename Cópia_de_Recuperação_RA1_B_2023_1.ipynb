#Questão 1
A seguir estão listados três limites. Sua tarefa será encontar o valor destes limites de duas formas: algébricamente e gráficamente.

a) Para a solução algébrica, você deve explicitar todas as passagens necessárias em latex, em uma célula de texto.

b) Para solucionar graficamente deverá usar uma tabela de aproximação pela esquerda e pela direita. Lembre-se de plotar o gráfico, os pontos usados para a aproximação e de imprimir a tabela no terminal. 


1. $$\lim_{{x \to 0}} \frac{1 - \cos(2x)}{x^2}$$
---
2. $$\lim_{{x \to \infty}} \frac{3x^2 - 2x + 1}{2x^2 + x - 1}$$
---
3. $$\lim_{{x \to 0}} x \cdot \ln(1 + \sin(\frac{1}{x}))$$


Respostas:
1. $$\lim_{{x \to 0}} \frac{1 - \cos(2x)}{x^2}$$
Aplica L’Hospital’s Rule:
$$\lim_{{x \to 0}} \frac{2\sin(2x)}{2x}=\frac{4\cos(2x)}{2}=\frac{4\cos(0)}{2}=\frac{4}{2}=2$$
---
2. $$\lim_{{x \to \infty}} \frac{3x^2 - 2x + 1}{2x^2 + x - 1}=\frac{3x^2}{2x^2}=\frac{3}{2}$$
---
3. Qualquer coisa vezes 0 é 0$$\lim_{{x \to 0}} x \cdot \ln(1 + \sin(\frac{1}{x}))=0$$

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
  return (1-np.cos(2*x))/x**2

def calcular_limites(x0, num_pontos, xmin, xmax):
    valores_esquerda = np.linspace(xmin, x0, num_pontos, endpoint=False)
    valores_direita = np.linspace(x0, xmax, num_pontos, endpoint=False)

    tabela_esquerda = [(x, f(x)) for x in valores_esquerda]
    tabela_direita = [(x, f(x)) for x in valores_direita]

    return tabela_esquerda, tabela_direita

def plotar_grafico(tabela_esquerda, tabela_direita, x0, xmin, xmax):
    plt.figure(figsize=(10, 6))

    x = np.linspace(xmin, xmax, 1000)
    plt.plot(x, f(x), label="f(x) = (1-cos(2x))/x^2")

    for x, y in tabela_esquerda:
        plt.plot(x, y, 'ro')
        plt.annotate(f'({x:.2f}, {y:.2f})', (x, y), textcoords="offset points", xytext=(-15, 5), fontsize=10, color='red')

    for x, y in tabela_direita:
        plt.plot(x, y, 'bo')
        plt.annotate(f'({x:.2f}, {y:.2f})', (x, y), textcoords="offset points", xytext=(-15, 5), fontsize=10, color='blue')

    plt.axvline(x=x0, color='g', linestyle='--', label="x = 0")
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.show()

x0 = 0
num_pontos = 4
xmin = -0.2
xmax = 0.2
tabela_esquerda, tabela_direita = calcular_limites(x0, num_pontos, xmin, xmax)
print("Tabela de valores à esquerda:")
print(tabela_esquerda)
print("\nTabela de valores à direita:")
print(tabela_direita)
plotar_grafico(tabela_esquerda, tabela_direita, x0, xmin, xmax)
```
